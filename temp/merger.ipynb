```python
import pandas as pd 
```

```python
file_template = '第四十期形势与政策网上考试-3018233061-樊一飞-数学学院 ({}).xls'
```

```python
df1 = pd.read_excel(file_template.format(2),'单选')
df2 = pd.read_excel(file_template.format(1),'单选')
```

```python
def read(i,section):
    return pd.read_excel(file_template.format(i),section)
```

```python
def merge(section,amount):
    for i in range(amount):
        df = read(i,section) 
        lib = df if i is 0 else pd.merge(lib,df,how='outer')
    return lib   
```

```python
df_sc = merge('单选',27)
df_sc
```

```python
df_mc = merge('多选',27)
df_mc
```

```python
writer = pd.ExcelWriter('lib.xlsx')
df_sc.to_excel(writer,sheet_name='单选',index=False)
df_mc.to_excel(writer,sheet_name='多选',index=False)
writer.save()
writer.close()
```
